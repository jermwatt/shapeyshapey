```python
import os
dir_path = os.path.abspath(os.getcwd())

# load in parent directory of dir_path
parent_dir = os.path.dirname(dir_path)

# add to path
import sys
sys.path.append(parent_dir)

# add path to library
library_path = parent_dir + '/shape_generator'
sys.path.append(library_path)

# add path to data output
data_path = parent_dir + '/shape_dataset'

# load in modules
import numpy as np 
from shape_generator.simple_square_tiling import create_complete_tiling, display_random_instance, create_gif, save_images_as_idx, get_random_image_pairs, train_test_split_image_pairs, read_all_image_pairs_from_idx

%load_ext autoreload
%autoreload 2
```

```python
# create tiling pairs
tilings = create_complete_tiling(img_width=32, square_width=5)
tiling_pairs = get_random_image_pairs(tilings)
train_set, test_set = train_test_split_image_pairs(tiling_pairs, test_ratio=0.1)

# save tiling pairs as idx files
train_save_path = data_path + "/train-shape-images-idx3-ubyte"
test_save_path = data_path + "/test-shape-images-idx3-ubyte"
save_images_as_idx(train_set, train_save_path)
save_images_as_idx(test_set, test_save_path)

# report
print('type of tiing', type(tilings[0]))
print('tiling shape', tilings[0].shape)
print('number of tiles', len(tilings))

print ('length of get_random_image_pairs', len(tiling_pairs))
print('length of training set', len(train_set))
print('length of test set', len(test_set))
```
