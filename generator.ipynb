```python
import struct
import numpy as np

def save_images(images, file_path):
    num_images = len(images)
    image_size = images[0].shape[0]

    # Open the file in binary mode for writing
    with open(file_path, 'wb') as f:
        # Write the IDX file header
        f.write(struct.pack('>H', 0))  # Magic number: 0 (for images)
        f.write(struct.pack('>H', num_images))
        f.write(struct.pack('>H', image_size))
        f.write(struct.pack('>H', image_size))

        # Write each image
        for image in images:
            # Convert image data to bytes
            image_data = image.astype(np.uint8).tobytes()
            f.write(image_data)
```

```python
file_path = 'test_images.idx'
save_images(tilings,file_path)
```
